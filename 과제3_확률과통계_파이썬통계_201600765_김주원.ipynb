### 2. 다음의 데이터를 이용함: 2 9 8 12 7 5 14 3 11 17
a. 데이터를 벡터 x에 저장


b. y = x^2 - 3 을 구함


c. x와 y의 correlation과 covariance를 구함


```python
# 문제 2 : numpy 라이브러리 사용

#a. 데이터를 벡터 x에 저장
x = np.array([2, 9, 8, 2, 7, 5, 14, 3, 11, 17])
print("a. 데이터를 x에 저장:", x)

#b. y = x^2 - 3 을 구함
y = np.power(x,2) - 3
print("b. y = x^2 - 3 :", y)

#c. x와 y의 correlation과 covariance를 구함
print("c")
print("- x,y의 피어슨 상관계수 계산 행렬:",np.corrcoef(x,y))
print("- x,y의 상관계수:",np.corrcoef(x,y)[0,1])
print("- x,y 의 cross-correlation: ", np.correlate(x,y))
print("- x,y의 공분산 행렬:",np.cov(x,y))
print("- x,y의 공분산:",np.cov(x,y)[0,1])
```

### 3. (-10, 20) 구간에서 다음 함수의 그래프를 그림

### 4. [-3, 5] 구간의 균일 분포의 난수 1000개를 생성하여 x에 저장

a. x의 평균과 표준편차를 구함

b. 0.25 간격으로 히스토그램을 구하고 plot 함


```python
# 4. numpy, matplotlib 이용
# [-3,5]구간의 균일분포, 난수 1000개 생성
x = np.random.uniform(-3,5,1000)

# a.평균과 표준편차 계산
print("a.평균:", np.mean(x),"표준편차:",np.std(x))

# b. 0.25간격의 히스토그램
weight,bins=np.histogram(x,bins=[-3,-2.75,-2.5,-2.25,-2,-1.75,-1.5,
                                 -1.25,-1,-0.75,-0.5,-0.25,0,0.25,0.5,0.75,1,
                                 1.25,1.5,1.75,2,2.25,2.5,2.75,3,3.25,3.5,3.75,4,
                                 4.25,4.5,4.75,5])
print("b.히스토그램:",weight)
print("구간", bins)
plt.title("0.25간격의 히스토그램")
plt.grid()
plt.hist(x, bins=bins, color = 'skyblue')
plt.ylabel('y축')
plt.xlabel('x축')
```

### 5. 평균이 7인 포아송 분포의 난수 500개를 생성하여 x에 저장
a. x의 평균과 표준편차를 구함

b.히스토그램을 구하고 plot 함


```python
# 5. numpy, matplotlib의 plt를 사용

x = np.random.poisson(lam=7, size=500)

# 평균과 표준편차 계싼
print("a.평균:", np.mean(x),"표준편차:",np.std(x))

# 0.25간격의 히스토그램
weight,bins=np.histogram(x)
print("b.히스토그램:",weight)
print("구간", bins)
plt.title("포아송 히스토그램")
plt.grid()
plt.hist(x, bins=bins, color = 'skyblue')
plt.ylabel('y축')
plt.xlabel('x축')
```

### 6. 주사위를 30번 던졌을 때 3이 5번 이상 나올 확률을 구함


```python
# 6. stats이용
# 주사위를 30번 던졌을 때 3이 5번 이상 나올 확율은 전체 확률에서 30번 던졌을때 3이 5번 미만인 확률을 빼주면 된다.

#확률 P
P = 1- (stats.binom.pmf(k=4, n=30, p=1/6)+stats.binom.pmf(k=3, n=30, p=1/6) + stats.binom.pmf(k=2, n=30, p=1/6)+ 
    stats.binom.pmf(k=1, n=30, p=1/6)+stats.binom.pmf(k=0, n=30, p=1/6))
print("정답:",P)
```

### 7. 평균이 20, 표준편차가 3인 정규분포의 랜덤변수 X에 대해 다음을 만족시키는 a를 구함


```python
# 7. stats 이용
# 평균=20, 표준편차=3 정규분포의 랜덤변수 X 생성
X = stats.norm(loc=20, scale=3)

# P(x>=a)=0.3 => 1 - P(x<a)=0.3인 부분과 같으므로 P(x<a)=0.7인 부분을 찾아서 구해주면 된다. 즉, 다시 말해서 0.7이 되는
# 누적 함수 구간의 a값을 구해주면 되고, 이 곳에서는 ppf(inversion cumulative distribution)을 이용해서 구해주면된다.
a= X.ppf(0.7)
print("a:",a)
```

### 8. 평균이 0인 정규분포를 따르는 잡음을 Gaussian noise 라고 하는데, 여기서는 이러한 잡음 특성을 가지는 신호를 생성하는 실험을 수행함

a. 신호 성분으로 다음을 생성.  x의 간격은 0.01로 함 

b. 평균이 0, 표준편차가 s인 난수를 x의 개수 만큼 생성하여 n에 저장

c. y=s + n 에 의해 잡음이 섞인 신호를 생성하여 그래프를 그림

d. 	s = 1, 2, 3, 4, 5로 변화시키면서 5개의 그래프를 그림

```python
# 7.numpy, matplotlib사용 
# a.
print("a.") 
# 정의역 [0,3]생성, 0.01단위
x = np.linspace(0,3,300)
s = 10* np.sin(2*np.pi *x)
plt.title("s = 10*sin(2pix)함수계형")
plt.plot(x,s)
plt.xlabel('x축')
plt.ylabel('s축')
```

```python
# b. 표준편차의 값이 문제에 정해져있지 않으므로 평균이 0인 표준 정규분포를 이용해서 생성
print("b. 평균이 0, 표준편차 sigma인  난수를 x의 개수 만큼 생성하여 n에 저장")
sigma = 1
n = np.random.normal(0.0, sigma, size=300)
print("n:",n)
y = s + n
plt.title("y= s+n함수계형")
plt.plot(x,y)
plt.xlabel('x축')
plt.ylabel('y축')
```

```python
print("d. sigma = 1, 2, 3, 4, 5로 변화시키면서 5개의 그래프를 그림")

#sigma = 1일떄
sigma = 1
n = np.random.normal(0.0, sigma, size=300)
y = s + n
plt.title("y= s+n함수계형(sigma=1) ")
plt.plot(x,y, color='skyblue')
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()

#sigma = 2일떄
sigma = 2
n = np.random.normal(0.0, sigma, size=300)
y = s + n
plt.title("y= s+n함수계형(sigma=2) ")
plt.plot(x,y, color='brown')
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()

#sigma = 3일떄
sigma = 3
n = np.random.normal(0.0, sigma, size=300)
y = s + n
plt.title("y= s+n함수계형(sigma=3) ")
plt.plot(x,y, color='red')
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()

#sigma = 4일떄
sigma = 5
n = np.random.normal(0.0, sigma, size=300)
y = s + n
plt.title("y= s+n함수계형(sigma=4) ")
plt.plot(x,y, color='limegreen')
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()

#sigma = 5일떄
sigma = 5
n = np.random.normal(0.0, sigma, size=300)
y = s + n
plt.title("y= s+n함수계형(sigma=5) ")
plt.plot(x,y, color='green')
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()
```
